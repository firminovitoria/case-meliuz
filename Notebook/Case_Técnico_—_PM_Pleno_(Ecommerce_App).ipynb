# Extração dos parâmetros customizados

```python
import json
import pandas as pd

df["tracking_json"] = df["tracking_url_params"].apply(
    lambda x: json.loads(x) if pd.notnull(x) else {}
)
```

```python
df["variant_raw"] = df["tracking_json"].apply(
    lambda x: x.get("mz_test_gotoexternalbrowser")
)

df["redirect"] = df["tracking_json"].apply(
    lambda x: x.get("mz_redirect")
)
```

Identificação das variantes

```python
variant_map = {
    "a": "Controle",
    "b": "Header",
    "c": "Config"
}

df["variant"] = df["variant_raw"].map(variant_map)
```

```python
df["variant"].value_counts(dropna=False)
```

Criação do tipo de saída

```python
def classify_exit(row):

    utm_content = str(row["utm_content"]).lower()
    utm_term = str(row["utm_term"]).lower()

    if utm_content == "partner_page":
        return "normal_inapp"

    elif utm_term == "header":
        return "external_header"

    elif utm_term == "login":
        return "external_login"

    elif utm_term == "config":
        return "external_config"

    else:
        return "other"

df["exit_type"] = df.apply(classify_exit, axis=1)
```

```python
df["exit_type"].value_counts()
```

Conversão

```python
df["converted"] = df["transaction_id"].notnull()
```

# Hipóteses do experimento


#### **Variante A — Controle**
*   **Hipótese:** Manter o usuário dentro do app maximiza o controle da jornada e preserva o tracking.

#### **Variante B — Header**
*   **Hipótese:** Facilitar a saída externa reduz a fricção e melhora a conversão.

#### **Variante C — Config**
*   **Hipótese:** Permitir a saída externa de forma menos intrusiva reduz fricções específicas sem degradar a jornada principal.

*   **Queda de Receita Média:** Alerta se a mudança canibalizar vendas de maior valor.
*   **Aumento Excessivo de Saída Externa:** Monitorar se o usuário está saindo e não retornando.

*   **Continuidade da Jornada:** Impacto negativo na retenção e navegação posterior do usuário.

**Validação estatística das diferenças de conversão**

```python
from statsmodels.stats.proportion import proportions_ztest

conversion_summary = df.groupby("variant")["converted"].agg(
    ["sum", "count"]
)

conversion_summary
```

```python
success = [
    conversion_summary.loc["Controle", "sum"],
    conversion_summary.loc["Header", "sum"]
]

nobs = [
    conversion_summary.loc["Controle", "count"],
    conversion_summary.loc["Header", "count"]
]

stat, pvalue = proportions_ztest(success, nobs)

print("Controle vs Header")
print("Z-stat:", stat)
print("P-value:", pvalue)
```

A comparação entre as variantes Controle e Header apresentou evidências de significância estatística `(p-value < 0.05)`, indicando que a diferença observada de conversão dificilmente ocorreu por acaso.
O resultado reforça a hipótese de que a exposição direta da funcionalidade de abertura externa no header impactou negativamente a jornada, aumentando mudanças de contexto e reduzindo a eficiência do fluxo de compra.

```python
success = [
    conversion_summary.loc["Controle", "sum"],
    conversion_summary.loc["Config", "sum"]
]

nobs = [
    conversion_summary.loc["Controle", "count"],
    conversion_summary.loc["Config", "count"]
]

stat, pvalue = proportions_ztest(success, nobs)

print("Controle vs Config")
print("Z-stat:", stat)
print("P-value:", pvalue)
```

A comparação entre as variantes Controle e Config não apresentou evidências de diferença estatisticamente significativa `(p-value > 0.05)`.  Isso sugere que permitir abertura externa de forma menos intrusiva, através do menu de configurações, preserva o desempenho agregado da jornada sem causar degradação relevante nas métricas de conversão. O resultado fortalece a interpretação de que a funcionalidade pode atuar como fallback contextual sem comprometer significativamente a experiência principal do In-App Browser.
